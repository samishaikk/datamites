### Datatime

```python
from datetime import datetime as dt 
```

```python
my_dt = dt(1990,12,3,6,32,54)
my_dt
```

```python
my_dt.year
```

```python
current_time = dt.now()
current_time
```

```python
current_time.weekday()
```

```python
from datetime import timedelta as td
```

```python
future_time = current_time + td(weeks=144)
future_time
```

```python
future_time.isocalendar()
```

```python
past_time = current_time - td(days = 5000)
past_time
```

```python
past_time.isocalendar()
```

### conditional statement
- if
- else 
- elif

```python
weather = 'rain'

if weather == "rain":
    print('wear raincoat')
else:
    print('do not')
```

```python
# nested if-else

num = int(input("enter the input: "))

if num > 50:
    if num > 90:
        print(f'output is {num + 30}')
    else:
        print(f'output is {num - 30}')
else:
    print(f'Given input is {num}')
```

### Loops

- for
- while

```python
list1 = [10,20,30,40,50]
list1
```

```python
for x in list1:
    print(x)
```

```python
for a in "Hello world":
    print(a.upper())
```

```python
list(range(10,40,5))
```

```python
for i in range(10):
    print(i+3)
```

```python
# nested for loop
list1
```

```python
list2 = [[1,2,3],[4,5,6]]
list2
```

```python
for a in list2:
    print(a)
    for b in a:
        print(b)
```

```python
# while loop

i = 1

while i<=5:
    print(i)
    i += 1
```

```python
# break 

num = 1
while num<=10:
    if num == 5:
        break
    print(num)
    num += 1
```

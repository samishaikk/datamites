```python
import matplotlib.pyplot as plt
```

```python
Overs = [3,11,15,17,22,44,46,47,48,49]
Runs = [2,33,44,55,22,66,77,22,77,44]
```

```python
plt.Figure(figsize=(4,3))
plt.plot(Runs,Overs)
plt.xlabel('Runs')
plt.ylabel('Overs')
plt.title('Runrate')
plt.show()
```

```python
plt.plot(Runs,Overs,marker='s',mfc='green',mec='yellow',linestyle='-.',color='black')
plt.show()
```

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.species.unique()
```

```python
df.species.value_counts()
```

```python
sns.lineplot(x='sepal_length',y='petal_width',data=df)
plt.Figure(figsize=(4,3))
plt.title('irisgraph')
plt.show()
```

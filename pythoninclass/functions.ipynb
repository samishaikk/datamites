```python
def basicfunc(num):
    if num>50:
        if num>80:
            print(num+30)
        else:
            print(num-30)
    else:
        print(num)
        
num = int(input())
basicfunc(num)
```

```python
def rect_area(length,breadth):
    area = length*breadth
    print(f'area of rect is: {area}')
    
length = int(input())
breadth = int(input())
rect_area(length,breadth)
```

### operators

- Arithmetic operators -> addition, sub, mult, division, floor division, modulo, exponential
- Assignment 
- Conditional
- Logical

```python
x = 33
y = 34
z = x+y
print(f"sum of {x} and {y} is {z}")
# print("sum of {} and {} is {}".format(x, y, z))
```

```python
str1 = "Data"
str2 = "science"
print(str1 +' '+ str2)
```

```python
d = x-y
P = x*y
division = y/x
floordivision = y//x
modulus = y % x
print(f"difference between {x} and {y} is {d}")
print(f'product of {x} and {y} is {P}')
print(f'division of {x} and {y} is {division}')
print(f'floor division of {x} and {y} is {floordivision}')
print(f'modulus of {x} and {y} is {modulus}')
```
